```python
import matplotlib as plt
import numpy as np
import pandas as pd
from scipy.spatial import Voronoi, voronoi_plot_2d
import folium
from folium.plugins import HeatMap
```

```python
df=pd.read_csv('/content/BPD_Arrests - Copy.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
cleaned_df = df.dropna(subset=['Latitude', 'Longitude', 'Gender'])
cleaned_df.head()
```

```python
males_df = cleaned_df[cleaned_df['Gender'] == 'M']
females_df = cleaned_df[cleaned_df['Gender'] == 'F']
```

```python

map_center = [39.299236, -76.609383]

combined_map = folium.Map(location=map_center, zoom_start=11)

heat_map_males = males_df[['Latitude', 'Longitude']].values
heat_map_females = females_df[['Latitude', 'Longitude']].values

heatmap_males_layer = HeatMap(heat_map_males, name="Males")
heatmap_males_layer.add_to(combined_map)

heatmap_females_layer = HeatMap(heat_map_females, name="Females")
heatmap_females_layer.add_to(combined_map)

folium.LayerControl().add_to(combined_map)

combined_map_path = "arrest_gender_heatmap.html"
combined_map.save(combined_map_path)

combined_map_path

```

It seems CDS (Controlled Dangerous Substance ) related crimes are the most frequent ones. So I create a map for arrests of males and females for CDS related crimes.

```python

# Filter the data for "CDS" offenses
cds_df = cleaned_df[cleaned_df['ChargeDescription'].str.contains('CDS', case=False, na=False)]

# Separate the filtered data into males and females
males_cds_df = cds_df[cds_df['Gender'] == 'M']
females_cds_df = cds_df[cds_df['Gender'] == 'F']


map_center = [39.299236, -76.609383]

combined_map = folium.Map(location=map_center, zoom_start=11)

heat_map_males_cds = males_cds_df[['Latitude', 'Longitude']].values
heat_map_females_cds = females_cds_df[['Latitude', 'Longitude']].values

# Create heatmap layer for males (for CDS offenses)
heatmap_males_layer = HeatMap(heat_map_males_cds, name="Males - CDS Offenses")
heatmap_males_layer.add_to(combined_map)

# Create heatmap layer for females (for CDS offenses)
heatmap_females_layer = HeatMap(heat_map_females_cds, name="Females - CDS Offenses")
heatmap_females_layer.add_to(combined_map)
folium.LayerControl().add_to(combined_map)

combined_map_path = "cds_gender_heatmap.html"
combined_map.save(combined_map_path)
combined_map_path
```
